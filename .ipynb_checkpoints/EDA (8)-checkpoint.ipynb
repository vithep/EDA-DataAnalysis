# Analise Exploratoria

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Juntando todos os arquivos

```python
files =[file for file in os.listdir("Dados/Sales_Data")]
for file in files:
    print(file)
```

```python
path = "Dados/Sales_Data"

all_data = pd.DataFrame()

for file in files:
    current_df = pd.read_csv(path+"/"+file)
    all_data = pd.concat([all_data, current_df])
    
all_data.shape
```

### Convertendo em um unico dataset

```python
all_data.to_csv('Dados/Sales_Data/all_data.csv',index=False)
```

```python
all_data.dtypes
```

```python
all_data.head()
```

### Verificando valores nulos

```python
x=all_data.isnull().sum()
y=(all_data.isnull().sum()/all_data.shape[0])*100
z={'Number of missing values':x,'Percentage of missing values':y}
df_1=pd.DataFrame(z,columns=['Number of missing values','Percentage of missing values'])
df_1.sort_values(by='Percentage of missing values',ascending=False)
```

```python
all_data = all_data.dropna(how='all')
all_data.shape
```

### Qual o melhor mÃªs para venda?

```python
'04/19/19 08:46'.split('/')[0]
```

```python
def month(x):
    return x.split('/')[0]
```

```python
all_data['Month']=all_data['Order Date'].apply(month)
```

```python
all_data.dtypes
```

```python
all_data['Month']=all_data['Month'].astype(int)
```

```python
all_data['Month'].unique()
```

```python
filter=all_data['Month']=='Order Date'
len(all_data[~filter])
```

```python
all_data=all_data[~filter]
```

```python
all_data.shape
```

```python
all_data.head()
```

```python
all_data['Month']=all_data['Month'].astype(int)
```

```python
all_data.dtypes
```

```python
all_data['Price Each']=all_data['Price Each'].astype(float)
```

```python
all_data['Quantity Ordered']=all_data['Quantity Ordered'].astype(int)
```

```python
all_data['sales']=all_data['Quantity Ordered']*all_data['Price Each']
all_data.head(5)
```

```python
all_data.groupby('Month')['sales'].sum()
```

```python
# configurar a visualizaÃ§Ã£o
#%matplotlib inline
%config InlineBackend.figure_format = 'svg'
#configurando parÃ¢metros das fontes
plt.rcParams['font.family'] = 'sans-serif'
```

```python
plt.figure(figsize=(12,6))
months=range(1,13)
plt.bar(months,all_data.groupby('Month')['sales'].sum())
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number');
```

### Qual cidade tem mais pedidos?

```python
'917 1st St, Dallas, TX 75001'.split(',')[1]
```

```python
def city(x):
    return x.split(',')[1]
```

```python
all_data['city']=all_data['Purchase Address'].apply(city)
```

```python
all_data.groupby('city')['city'].count()
```

```python
plt.figure(figsize=(11,6))
plt.bar(all_data.groupby('city')['city'].count().index,all_data.groupby('city')['city'].count())
plt.xticks(rotation='vertical')
plt.ylabel('received orders')
plt.xlabel('city names');
```

### Qual horario devemos enviar avisos para potencializar as vendas

```python
all_data['Hour'] = pd.to_datetime(all_data['Order Date']).dt.hour
```

```python
keys=[]
hour=[]
for key,hour_df in all_data.groupby('Hour'):
    keys.append(key)
    hour.append(len(hour_df))
```

```python
plt.grid()
plt.plot(keys,hour);
```

### Qual produto Ã© mais vendido e porque?

```python
all_data.groupby('Product')['Quantity Ordered'].sum().plot(kind='bar')
```

```python
all_data.groupby('Product')['Price Each'].mean()
```

```python
products=all_data.groupby('Product')['Quantity Ordered'].sum().index
quantity=all_data.groupby('Product')['Quantity Ordered'].sum()
prices=all_data.groupby('Product')['Price Each'].mean()
```

```python
plt.figure(figsize=(11,6))
fig,ax1 = plt.subplots()
ax2=ax1.twinx()
ax1.bar(products, quantity, color='g')
ax2.plot(products, prices, 'b-')
ax1.set_xticklabels(products, rotation='vertical', size=8)
```
